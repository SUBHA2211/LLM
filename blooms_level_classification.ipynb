```python
# Install necessary libraries
!pip install transformers openpyxl pandas

# Step 1: Import Libraries
import pandas as pd
from transformers import pipeline

# Step 2: Load Your Excel File
# Replace with your filename if it's different
file_path = "/content/new_final (1).xlsx"
df = pd.read_excel(file_path)

# Preview the data
df.head()
```

```python
df = df.head(100)  # or use df.iloc[:100]

# Step 4: Preview the Selected Data
print("First 100 Rows:")
print(df.head())
```

```python
df.size
```

```python
# Define Bloom's levels as candidate labels
candidate_labels = ["Remember", "Understand", "Apply", "Analyze", "Evaluate", "Create"]
```

```python
from transformers import pipeline

# Load the zero-shot classifier
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli",device ="cuda")  # or try: "MoritzLaurer/deberta-v3-large-zeroshot-v2"
```

```python
# Define a function to classify Bloom's level
def classify_blooms(question):
  if isinstance(question, str):  # Check if it's a string
    result = classifier(question, candidate_labels, multi_label=False)
    return result['labels'][0]  # Top predicted label
  else:
    return None  # Return None for non-string values


# Apply to the DataFrame
df['Bloom_Level'] = df['Question'].apply(classify_blooms)
```

```python
output_path = "/content/questions_with_blooms.xlsx"
df.to_excel(output_path, index=False)

print(f"Saved with Bloom levels at: {output_path}")
```
